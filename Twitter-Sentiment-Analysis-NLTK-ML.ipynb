>**Sentiment analysis** (emotion AI) is the process of “computationally identifying and categorizing opinions expressed in a piece of text… to determine whether the writer's attitude towards a topic, product is positive, negative, or neutral” [1].

Sentiment analysis is a form of text analytics that uses Natural Language Processing (NLP) in combination with machine learning (ML) to interpret emotions in subjective data. There are two main forms of sentiment analysis: lexicon or rule based and automated or machine learning based; this investigation focuses on the latter. This process of "opinion mining" enables polarity classification of a collection of texts and facilitates the interpretation of the public sentiment towards Musk. 

>**Natural Language Processing (NLP)** is a sub-field of computer science and artificial intelligence which focuses on programming computers to process and analyze textual data. 

NLP techniques are often used for sentiment analysis because they can help to analyze and understand the structure and meaning of human language.

There are various ways to perform sentiment analysis using NLP techniques, such as:

> **Lexical analysis**: This involves analyzing the words and phrases used in a text to determine its sentiment. For example, words like "excellent" and "wonderful" are generally considered positive, while words like "terrible" and "awful" are considered negative.  


> **Syntactic analysis**: This involves analyzing the grammar and syntax of a text to understand its meaning and sentiment. For example, the use of negation words like "not" and "never" can reverse the sentiment of a phrase.  


> **Semantic analysis**: This involves analyzing the meaning and context of words and phrases in a text to understand its overall sentiment. For example, the phrase "I am so tired" may be interpreted as negative, while the phrase "I am so excited" may be interpreted as positive.  

NLP techniques are useful for sentiment analysis because they can help to analyze and understand the structure and meaning of human language, which is crucial for accurately extracting sentiment from text data. The focus of this investigation will be in relation to lexical and semantic analysis, and not syntactic analysis.

## 1.2 Justification of Data Source

### 1.2.1 Data Source Origin and Acquisition Techniques

The Twitter database is composed of a collection of Tweets. Each Tweet is a short message, typically consisting of up to 280 characters, that is posted on the social media platform Twitter. Tweets can include text, hashtags, images, videos, and links to other web pages. They are created and posted by users of the platform, and they can be viewed by anyone who follows the user who posted the Tweet. The Twitter community generates more than 12 terabytes of data per day [2]. Twitter is composed of clusters storing over 500 PB of data, with the largest cluster containing 10K nodes. Twitter uses distributed storage using the Gizzard framework to speed up data processing [3]. Gizzard is built on top of MySQL (InnoDB); InnoDB was selected because it does not corrupt data. The Tweet data is stored in an internal system called T-bird, which is built on top of Gizzard. Secondary indexes are stored in a Gizzard based system known as T-flock.

Snscrape is a Python library that can be used as a scraper for social networking services (SNS). It is able to scrape Twitter users, user profiles, hashtags, searches, Tweets (single or surrounding thread), trends, and list posts. It is selected because it can scrape Tweets through Twitter’s API without request limits or restrictions. Moreover, a Twitter developer account is not required.

```python
# Function that removes URLs
def remove_URLs(tweet_body):
     return re.sub(r"http\S+", "", tweet_body)
        
# Remove URLs from dataset
data['Tweet'] = data['Tweet'].apply(lambda x: remove_URLs(x))
```

### 2.2.6 Remove Punctuations that Delimit Words

```python
# String constant containing all sets of punctuation
english_punctuation = '''()-[]{};:'"\,<>./…’?@#$%^&!*_~’'“’”'''

# Function that removes English punctuation
def remove_english_punctuation(tweet_body):
    
    # Create a one-to-one mapping table to map punctuation to None
    translation_table = str.maketrans('', '', english_punctuation)
    
    return tweet_body.translate(translation_table)

# Remove punctuation
data['Tweet'] = data['Tweet'].apply(lambda x: remove_english_punctuation(x))
```

```python
# Remove excess whitespace from Tweets
data['Tweet'] = data['Tweet'].apply(lambda x: re.sub(' +', ' ', x))
```

### 2.2.8 Remove Emojis (Emoticons, Symbols and Pictographs, Transport and Map Symbols, Dingbats, etc.)

```python
# Function to remove emojis
def remove_emojis(tweet_body):
    
        # Set of Unicode emojis to remove
        emoji_pattern = re.compile("["
                                   
        # Emoticons
        u"\U0001F600-\U0001F64F"  
                                   
        # Symbols and pictographs      
        u"\U0001F300-\U0001F5FF"  
                                   
        # Transport and map symbols
        u"\U0001F680-\U0001F6FF"  
                                   
        # Flags in iOS
        u"\U0001F1E0-\U0001F1FF"  
                                   
        # Chinese characters
        u"\U00002500-\U00002BEF" 
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
                                   
        # Dingbats
        u"\ufe0f"  
        u"\u3030""]+", re.UNICODE)
        
        return emoji_pattern.sub(r'', tweet_body) 
    
# Remove emojis from Tweets
data['Tweet'] = data['Tweet'].apply(lambda x : remove_emojis(x))
```

### 2.2.9 Remove Numeric Characters

### 2.2.10 Remove Out-of-bounds Characters Outside the Unicode Character Set

Filtering strings with non-Unicode characters ensures that the polarity of tokens remain decipherable.

## 2.4 Text Normalization for Improved Matching: Stemming and Lemmatization

>**Normalization** must be performed to convert a word into its canonical form. 

Each word can be represented in different forms (i.e. "swims", "swam" and "swimming" should be converted to the same canonical form, "swim") which must be normalized. For our research question, the variant forms of the same word should be treated as bearing the same meaning and sentiment through reduction to a common stem. This reduces the complexity and dimensionality of data (by removing inflections and variations), and thus improves word matching. Stemming and lemmatization are used as techniques for normalization. Stemming is a heuristic process that removes affixes from a word while lemmatization normalizes a word using morphological analysis and the context of vocabulary.

### 2.4.1 Stemming

>**Stemming** is used as a standardization technique. Normalizing text by transforming words to their canonical form is critical to building a robust model. This removes repetition faster than lemmatization, but does not take into account contextual cues.

The Porter Stemmer or Porter algorithm uses a five-step word reduction with each step containing a set of mapping rules [9]. The Snowball Stemmer is more precise and logical than the Porter Stemmer. The Lancaster Stemmer often results in over-stemming, which creates non-linguistic or meaningless terms. Thus, the Snowball Stemmer is selected, as it is universally accepted as being better than the Porter Stemmer, which is susceptible to under-stemming.

Formally, the common rules in Snowball Stemming include:

| Suffix | Replacement Suffix |
|:-------|:-------------------|
| ILY    | ILI                |
| LY     | Nill               |
| SS     | SS                 |
| S      | Nill               |
| ED     | E, Nill            |

```python
# Create instance of Snowball stemming algorithm to stem words to their root form
snowball_stemmer = nltk.SnowballStemmer('english')

# Function that performs stemming on Tweet
def stemming_normalization(tweet_body):
    stemmed_text = [snowball_stemmer.stem(word) for word in tweet_body]
    return stemmed_text

# Perform stemming on Tweets
data['Tweet']= data['Tweet'].apply(lambda x: stemming_normalization(x))
```

### 2.4.2 Lemmatization
>**Lemmatization** enables syntactical normalization (i.e. part of speech remains the same). This uses the concept of a lexeme. A word's meaning is based on its part of speech, which can elucidate different lexemes.

```python
# Download and import resources for lemmatization
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer

# Create instance of lexical database that provides semantic relationships between its words
lemmatizer = nltk.WordNetLemmatizer()

# Function that lemmatizes Tweet
def lemmatizer_normalization(tweet_body):
    lemmatized_text = [lemmatizer.lemmatize(word) for word in tweet_body]
    return lemmatized_text

# Perform lemmatization on Tweets
data['Tweet'] = data['Tweet'].apply(lambda x: lemmatizer_normalization(x))
```

Raw scores above the mean have positive standard scores whilst those below the mean have negative standard scores. This representation is provides informative descriptive power for skewed distributions.

Z-scores are able to identify outliers in a dataset. An outlier is a value that is significantly different from the other values in the dataset. Z-scores can help identify outliers because they provide a standardized way of comparing values across different datasets with different means and standard deviations. 

For example, a Z-score polarity of 4.0 indicates that the corresponding polarity value is 4 standard deviations above the mean, while a Z-score subjectivity of -0.7 indicates that the corresponding subjectivity value is 0.7 standard deviations below the mean. The former would suggest a major outlier.

```python
# Function to compute Z-scores (restandardized values) of a column to better characterize the column distributions

# Acquire the z-transform of a column by rescaling the data with mean = 0 and standard deviation = 1
def compute_standard_score(values):
    return (values - np.mean(values))/(np.std(values))
```

```python
# Iterate over DataFrame columns
for column in sanitizedTweets.columns:
    # Compute the polarity Z-scores
    if column == 'Polarity':
        sanitizedTweets['Z-Score Polarity'] = compute_standard_score(sanitizedTweets[column])
        
    # Compute the subjectivity Z-scores
    if column == 'Subjectivity':
        sanitizedTweets['Z-Score Subjectivity'] = compute_standard_score(sanitizedTweets[column])

# Visualize the DataFrame with Z-score columns
sanitizedTweets
```

## 3.5 Replace Polarity Value Outliers that Exceed Three Standard Deviations 

Outliers that contain polarity classifications more than three standard deviations away from the mean are replaced with the mean polarity value of the column. Since outliers deviate significantly from the norm, they impact the accuracy of the dataset. Modifying outliers or anomalies from a dataset prior to model training improves the performance and prediction accuracy of the model. Moreover, this improves interpretability by removing misleading or abnormal data points. Modifying outliers  also improves accuracy and reduces bias by creating a more representative dataset without the presence of anomalies. Outliers are transformed to the mean polarity value in order to make the data more normally distributed. Only outliers that fall outside of 99.9% of data points are modified; this corresponds to values within three standard deviations of the mean.

### 3.5.1 Rationale for using the Three Standard Deviations Threshold

In statistics, data points that are more than three standard deviations away from the mean are considered outliers. This is because, according to the empirical rule (also known as the $68 - 95 - 99.7$ rule), approximately 68% of the data points in a normal distribution fall within one standard deviation of the mean, approximately 95% of the data points fall within two standard deviations of the mean, and approximately 99.7% of the data points fall within three standard deviations of the mean. Therefore, if you have a dataset that is approximately normally distributed, you can use the three standard deviations rule to identify potential outliers in the data. 

### 3.5.2 Actual versus Predicted Number of Polarity Outliers Identified

867 outliers were found out of a total dataset of 30 000. Since 867/30 000 = 0.0289, 97.11% of the data fell within 3 standard deviations of the mean. This is in contrast to the predicted percentile of 99.7%. Thus, 2.59% more outliers were detected than expected. This suggests that the replacement of outliers will improve the accuracy of the mean and median values more than that of a standard dataset.

```python
# Compute the mean of the polarity column
polarityMean = np.mean(sanitizedTweets['Polarity'])

print("The mean polarity of the Tweets is: {:.4f}".format(polarityMean))

# Compute the standard deviation of the polarity column
polarityStandardDeviation = np.std(sanitizedTweets['Polarity'])
print("The standard deviation of polarity of the Tweets is: {:.4f}\n".format(polarityStandardDeviation))

# Variable to keep track of number of replacements
numReplacements = 0

# Replace polarity outliers that reside more than three standard deviations away from the mean with the value of the mean
for (index, value) in enumerate(sanitizedTweets['Polarity']):
    if value < (polarityMean - (polarityStandardDeviation * 3.0)) or value > (polarityMean + (polarityStandardDeviation * 3.0)):
        # Increment outlier replacement counter
        numReplacements += 1
        if numReplacements < 10:
            print("Replaced a polarity outlier at index {:.0f} with value: {:.2f}".format(index, value))
            value = polarityMean
        else:
            pass
        
print(".\n.\n.\n")
print("Replaced", numReplacements, "outliers in the dataset with mean polarity values.")
print(sanitizedTweets.shape[0] - numReplacements, "data points were within three standard deviations (~99.7% of the dataset) of the mean.")
```

### 4.2.1 Filter Out Top 15 Most Frequently Used Unique Words in Tweets and Plot in a Frequency Distribution

## 4.6 Converting Data into Appropriate Format for ML Pipeline using TF-IDF Vectorizer

TfidfVectorizer converts a collection of raw documents into a matrix of Term Frequency/Inverse Document Frequency (TF-IDF) features. The frequency of each word in the collection is used to compute its TF-IDF feature. It models the importance of terms within documents and converts text data into numerical form, which enables any machine learning or deep learning models to be used.

### 4.6.1 Term Frequency (TF)

The term frequency is the relative frequency of term $t$ in document $d$:  

$${\displaystyle \mathrm {tf} (t,d)={\frac {f_{t,d}}{\sum _{t'\in d}{f_{t',d}}}}},$$  

Where the numerator is the raw count of a term in a document.

### 4.6.2 Inverse Document Frequency (IDF)

The inverse document frequency is a measure of how common or rare a word is across different documents:

$$\mathrm{idf}(t, D) =  \log \frac{N}{|\{d \in D: t \in d\}|}$$

Where

* ${\displaystyle N}$: total number of documents in the corpus.  


* ${\displaystyle |\{d\in D:t\in d\}|}$: number of documents where the term ${\displaystyle t}$ is present. It is common to adjust the denominator to ${\displaystyle 1+|\{d\in D:t\in d\}|}$ to avoid a division-by-zero error when the term is not present in the corpus.

IDF is the logarithmically scaled inverse fraction of the documents containing the word (i.e. dividing the total number of documents by the number of documents that contain the term, and then taking the logarithm of the quotient) [10].

## 5.2 Evaluating Models using F1-Score: Precision and Recall

F1-Scores are a common metric for evaluating the performance of a classification model. The following key benefits support the usage of F1-Scores:

>**F1-Scores** account for both precision and recall: F1-Scores are a harmonic mean of precision and recall, which are two important evaluation metrics for classification tasks. By combining both metrics, F1 scores provide a more comprehensive evaluation of the model's performance.

>**Precision** measures the percentage of correct positive predictions among all positive predictions, and is computed dividing the true positives by any data that was predicted as a positive.     

>**Recall** (i.e. True Positive Rate) measures the percentage of correct positive predictions among all actual positive samples, and is computed by dividing the true positives by any data that should have been predicted as positive. 

* F1-Scores are not sensitive to class imbalance: In classification tasks, the classes may be imbalanced, meaning that one class may have significantly more samples than the other. This can make evaluation metrics like accuracy and precision less reliable, as they may be skewed by the class imbalance. F1-Scores are less sensitive to class imbalance, as they take into account both precision and recall, which can be affected differently by class imbalance.    


* F1-Scores are interpretable: F1 scores are easy to interpret, as they are a single score that ranges from 0 to 1, with higher values indicating better performance (i.e. 1 is optimal). This makes provides a state-of-the-art and intuitive method of comparing the performance of different models.

F1-Score can be computed as follows:

$${\displaystyle F_{1}={\frac {2}{\mathrm {recall} ^{-1}+\mathrm {precision} ^{-1}}}=2{\frac {\mathrm {precision} \cdot \mathrm {recall} }{\mathrm {precision} +\mathrm {recall} }}={\frac {2\mathrm {tp} }{2\mathrm {tp} +\mathrm {fp} +\mathrm {fn} }}}$$

### 5.2.1.1 Macro Averaging versus Weighted Averaging

Before analyzing the F1-Scores it is important to distinguished between macro averaging and weighted averaging.

>**Macro averaging** is a method of averaging the scores of multiple classes by treating each class equally. The scores of all classes are summed and divided by the number of classes, regardless of the number of samples in each class. Macro averaging is useful when you want it is beneficial to give each class equal weight.

>**Weighted averaging** is a method of averaging the scores of multiple classes by weighting each class according to the number of samples in that class. The scores of the classes are weighted according to the number of samples in each class, and then summed and divided by the total number of samples. Weighted averaging is useful when it is beneficial to give more weight to classes with more samples.

In this case, weighted averaging will increase the weight of the neutral class and decrease the weight of the negative class for polarity.

## 5.5 Use One-Versus-Rest (OVR) Approach to Multiclass Classification for ROC AUC Curve Plotting

>The **ROC AUC** is a metric that evaluates the performance of a binary classification model by plotting the true positive rate (TPR) against the false positive rate (FPR) at different probability thresholds. The AUC (Area Under the Curve) is a measure of the model's performance, with a value between 0 and 1. A value of 0.5 represents a random classifier, while a value of 1 represents a perfect classifier.

In a multiclass classification problem, where there are more than two classes, the ROC curve can still be used to evaluate the performance of the classifier. There are two main approaches for doing this: one-versus-rest (OVR) and one-versus-one (OVO). Both methods are heuristic methods for using binary classification algorithms for multi-class classification.

### 5.5.1 One-Versus-Rest (OVR) Approach

>In the **OVR approach**, a separate classifier is trained for each class to predict whether an instance belongs to that class or the remaining classes grouped together as a single unit. The ROC curve for each classifier can then be plotted on the same graph and the area under the curve (AUC) can be calculated for each class. This approach can be useful for comparing the performance of the classifier on different classes and for identifying any potential imbalances in the classification performance.

For example, OVR can be used as follows:

1. Binary Classification Problem I: Positive vs. [Negative, Neutral]
2. Binary Classification Problem II: Negative vs. [Positive, Neutral]
3. Binary Classification Problem III: Neutral vs. [Positive, Negative]

One model must be constructed for each class. This may be problematic for slow models such as neural networks, multidimensional classes or large datasets consisting of millions of rows. Since these limitations do not affect our use case, the OVR approach is selected.

### 5.5.2 One-Versus-One (OVO) Approach

>In the **OVO approach**, a separate classifier is trained for every unique pairwise combination of classes. The ROC curve for each classifier can then be plotted on the same graph and the AUC can be calculated for each class. This approach can be more computationally expensive than the OVR approach, but it may provide more detailed information about the performance of the classifier on different classes. For a multiclass classification problem with three classes, the one-versus-rest (OVR) approach and the one-versus-one (OVO) approach will both be equally efficient in terms of the number of classifiers that need to be trained.

The ROC curve and AUC are useful metrics for evaluating the performance of multiclass classifiers because they allow you to compare the performance of the classifier on different classes and to summarize the overall performance of the classifier. 

### 5.5.3 Selecting the OVR Approach over the OVO Approach

The main advantage of the OVR approach is that it is more interpretable. By training a separate classifier for each class, the OVR approach provides a clear comparison of the performance of the classifier on each class. This can be useful for identifying any potential imbalances in the classification performance and for comparing the performance of the classifier on different classes.


### 5.5.2 Benefits and Drawbacks of using both F1-Score and ROC AUC

The main benefit of using both F1-Score and ROC AUC is that a more comprehensive assessment of predictive capability is achieved. The F1-Score is a balanced metric based on precision and recall of a multivariate classification problem, while the ROC AUC value is a threshold-dependent metric based on a binary classification problem. If the ROC AUC value is computed using an OVR approach, in which each class is compared to the rest of the classes combined, the multiclass classification problem can utilize ROC AUC binary classification. Using both metrics creates a more reliable set of metrics that are more likely to be accurate and generalizable. 

The reason why an amalgamation of the two techniques serves as a better metric is because each technique has advantages over the other. 

* The main advantage of using the F1 score is that it takes both precision and recall into account, whereas the ROC AUC (Receiver Operating Characteristic Area Under the Curve) only considers the true positive rate (also known as recall). Another advantage of using the F1 score is that it is more interpretable than the ROC AUC. The ROC AUC is a summary measure that reflects the overall performance of the classifier, but it can be difficult to understand what a specific ROC AUC value means in terms of the classifier's behavior. In contrast, the F1 score is a single value that can be easily interpreted and compared to other classifiers.  


* One advantage of using the ROC AUC (Receiver Operating Characteristic Area Under the Curve) over the F1 score is that the ROC AUC is a summary measure that reflects the overall performance of a classifier, whereas the F1 score is a single value that is based on the precision and recall of the classifier. This can be particularly useful when you want to have a more comprehensive understanding of the classifier's behavior, as the ROC AUC takes into account the true positive rate (also known as recall) at various thresholds. Another advantage of the ROC AUC is that it is less sensitive to class imbalance compared to the F1 score. 
    * The F1 score is defined as the harmonic mean of precision and recall, where precision is the number of true positives divided by the sum of true positives and false positives, and recall is the number of true positives divided by the sum of true positives and false negatives. The F1 score gives equal weight to both precision and recall, which can be problematic in situations where the classes are imbalanced. For example, if the positive class is much rarer than the negative class, the classifier may have a high recall but a low precision, leading to a low F1 score. Thus, this must be corrected using oversampling or undersampling to balance the classes. In contrast, the ROC AUC only considers the true positive rate, so it is not as affected by class imbalance.

### 5.6.1 Create a Function to Compute ROC AUC Curve using y_test and y_pred for a Model

```python
# Import necessary libraries
from sklearn.metrics import roc_curve, auc

def plot_ROC_AUC(y_test, y_pred, title):
    # Calculate the True Positive Rate (TPR) and False Positive Rate (FPR) for each classification threshold
    fpr, tpr, thresholds = roc_curve(y_test, y_pred)

    # Calculate the AUC
    roc_auc = auc(fpr, tpr)

    # Plot the ROC curve
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.title(title)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
```
